* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Pulling the data needed from proper dataset to calculate 
total_schools = len(school_data)
total_students = len(student_data)
total_budget = school_data.budget.sum()
avg_math = student_data.math_score.mean()
avg_reading = student_data.reading_score.mean()


```

```python
#creating variable for math scores greater or equal to 70 
mathScore = student_data.math_score >= 70
passingMath = len(student_data.loc[mathScore])

mpercentage = 100 * (passingMath / total_students)
mpercentage
```

```python
#creating variable for reading scores greater or equal to 70 
readingScore = student_data.reading_score >= 70
passingReading = len(student_data.loc[readingScore])

readingPerc = 100 * (passingReading / total_students)
readingPerc
```

```python
#creating variable for both scores combined to see who is passing both classes 
both = (student_data.reading_score >= 70) & (student_data.math_score >= 70)
passingBoth = len(student_data.loc[both])

percentageBoth = 100 * (passingBoth / total_students)
percentageBoth
```

```python
# creating an output for the data we extracted above to make it readable
dataframe = {
    'Total Number of Schools': total_schools,
    'Total Number of Students': total_students,
    'Total Budget': total_budget,
    'Average Math Score': avg_math,
    'Average Reading Score': avg_reading,
    '% of Students Passing Math': mpercentage,
    '% of Students Passing Reading': readingPerc,
    '% Passing overall': percentageBoth
}

outputdata = pd.DataFrame([dataframe])
outputdata
```

```python
# number of studens passing math class

passing_math = df.loc[df.math_score >= 70].groupby(["school_name", "type"]).student_name.count()
avg_math = 100 * (passing_math / tot_students)
avg_math 
```

```python
# number of students passing reading class

passing_reading = df.loc[df.reading_score >= 70].groupby(["school_name", "type"]).student_name.count()
avg_reading = 100 * (passing_reading / tot_students)
avg_reading
```

```python
# number of students passing both classes
# not sure about the error I am getting but I looked at examples and the syntax seems right

passing_both = df.loc[(df.reading_score >= 70) & (df.math_score >= 70)].groupby(["school_name", "type"]).student_name.count()
avg_passing_both = 100 * (passing_both / tot_students)
avg_passing_both
```
